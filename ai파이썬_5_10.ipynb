```python
import pandas as pd

# 데이터 생성
data = {
    '학생': ['john', 'alice', 'bob', 'emily'],
    '수학점수': [90, 87, 78, 92],
    '영어점수': [85, 91, 80, 85],
    '과학점수': [88, 89, 85, 82]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 수학점수가 가장 높은 학생의 정보 출력
highest_math_score = df[df['수학점수'] == df['수학점수'].max()]
print(highest_math_score)
```

```python
import pandas as pd

# 데이터 생성
data = {
    '학생': ['john', 'alice', 'bob', 'emily'],
    '수학점수': [90, 87, 78, 92],
    '영어점수': [85, 91, 80, 85],
    '과학점수': [88, 89, 85, 82]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 수학 점수가 90점 이상인 학생의 정보 출력
math_score_above_90 = df[df['수학점수'] >= 90]
print(math_score_above_90)
```
