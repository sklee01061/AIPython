```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
wine_load = load_wine()
wine = pd.DataFrame(wine_load.data, columns=wine_load.feature_names)
wine['Class'] = wine_load.target
wine['Class'] = wine['Class'].map({0:'class_0',1:'class_1',2:'class_2'})
import matplotlib.pyplot as plt
plt.boxplot(wine['color_intensity'], whis=1.5)
plt.title('color_intensity')
plt.show()
```

```python
import numpy as np

quartile_1, quartile_3 = np.percentile(wine['color_intensity'], [25, 75])
iqr = quartile_3 - quartile_1
upper_whis = quartile_3 + 1.5 * iqr
lower_whis = quartile_1 - 1.5 * iqr

outliers = wine['color_intensity'][(wine['color_intensity'] > upper_whis) | (wine['color_intensity'] < lower_whis)]
print(outliers)
print(outliers.index)
```

```python
drop_outliers = wine.drop(index=outliers.index)
print('outliers :', wine.shape)
print('Drop outliers :', drop_outliers.shape)
```

```python
wine.loc[outliers.index, 'color_intensity'] = wine['color_intensity'].dropna().mean()
wine.loc[outliers.index, 'color_intensity']
```

```python
import pandas as pd

# 1. 데이터 파일 읽어오기
iris_data = pd.read_csv('/content/drive/MyDrive/ai 파이썬 자료 2/iris.csv')

```

```python
# 3. 데이터의 기초 통계량 출력
print("\n기초 통계량:")
print(iris_data.describe())
```

```python
print("\n품종별 개수:")
print(iris_data['species'].value_counts())
```

```python
print("\n결측치 확인:")
print(iris_data.isnull().sum())
```

```python
print("\n중복 데이터 개수 확인:")
print(iris_data.duplicated().sum())
```

```python
# 중복 데이터 모두 확인하기
print("\n중복 데이터 모두 확인:")
print(iris_data[iris_data.duplicated()])
```

```python
# 중복 데이터 삭제하기
iris_data = iris_data.drop_duplicates()
```

```python
print("\n품종별 합계:")
print(iris_data.groupby('species').sum())
```

```python
print("\n품종별 평균:")
print(iris_data.groupby('species').mean())
```

```python
import matplotlib.pyplot as plt
iris_data.groupby('species')['sepal_length'].mean().plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
iris_data.hist(figsize=(10, 8))
plt.show()
```

```python
plt.scatter(x=iris_data['sepal_length'], y=iris_data['petal_length'])
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
# 4. 품종별 개수 구하기
print("\n품종별 개수:")
print(iris_data['species'].value_counts())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Histogram of Iris Features', fontsize=16)

for i, feature in enumerate(iris_data.columns[:-1]): # 마지막 열(species)은 제외
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    iris_data[feature].hist(ax=ax, bins=20)
    ax.set_title(feature)

for ax in axes.flatten():
    ax.grid(False)

plt.tight_layout()
plt.show()
```
