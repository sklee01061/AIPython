```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터프레임 생성
data = {
    '요일': ['월', '화', '수', '목', '금', '토', '일'],
    '공부시간(시간)': [2, 3, 1, 5, 4, 2, 1],
    '성적': [70, 75, 65, 85, 80, 70, 60]
}

df = pd.DataFrame(data)

# 2. 요일별 공부 시간 선 그래프
plt.figure(figsize=(10, 5))
plt.plot(df['요일'], df['공부시간(시간)'], marker='o', linestyle='-', color='b', label='공부시간')
plt.xlabel('요일')
plt.ylabel('공부시간 (시간)')
plt.title('요일별 공부 시간')
plt.legend()
plt.grid(True)
plt.show()

# 3. 요일별 성적 선 그래프
plt.figure(figsize=(10, 5))
plt.plot(df['요일'], df['성적'], marker='^', linestyle='-', color='r', label='성적')
plt.xlabel('요일')
plt.ylabel('성적')
plt.title('요일별 성적')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터프레임 생성
data = {
    '요일': ['월', '화', '수', '목', '금', '토', '일'],
    '공부시간(시간)': [2, 3, 1, 5, 4, 2, 1],
    '성적': [70, 75, 65, 85, 80, 70, 60]
}

df = pd.DataFrame(data)

# 요일별 공부 시간 선 그래프
plt.figure(figsize=(10, 5))
plt.plot(df['요일'], df['공부시간(시간)'], marker='o', linestyle='--', color='b', label='공부시간')
plt.xlabel('요일')
plt.ylabel('공부시간 (시간)')
plt.title('요일별 공부 시간')
plt.legend()
plt.grid(False)  # 격자선 제거
plt.show()

# 요일별 성적 선 그래프
plt.figure(figsize=(10, 5))
plt.plot(df['요일'], df['성적'], marker='^', linestyle='--', color='r', label='성적')
plt.xlabel('요일')
plt.ylabel('성적')
plt.title('요일별 성적')
plt.legend()
plt.grid(False)  # 격자선 제거
plt.show()
```
